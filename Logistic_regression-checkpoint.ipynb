```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification 
```

```python
## create the dataset
x,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
```

```python
pd.DataFrame(x)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
## model training
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
```

```python
logistic.fit(x_train,y_train)
```

```python
y_pred = logistic.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score= accuracy_score(y_test,y_pred)
print(score)
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

## hyperparameter turning and cross validation

```python
model = LogisticRegression()
penalty = ['l1','l2','elasticnet']
c_values = [100,10,1.0,0.01]
solver=['newton-cg','lbfgs','liblinear','sag','saga']
```

```python
params=dict(penalty = penalty,C=c_values,solver=solver)
```

```python
## grid search cv
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
grid = GridSearchCV(estimator = model,param_grid = params,scoring = 'accuracy',cv=cv,n_jobs = -1)
```

```python
grid
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(x_test)
```

```python
score= accuracy_score(y_test,y_pred)
print(score)
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

# Logistic Regression for imbalanced dataset

```python
## generate and plot a synthetic imbalanced classification dataset
from collections import Counter
from sklearn.datasets import make_classification
```

```python
## imbalanced dataset
x,y = make_classification(n_samples=1000,n_features=2,n_clusters_per_class=2,n_redundant=0,weights=[0.99],random_state=10)
```

```python
x
```

```python
Counter(y)
```

```python
sns.scatterplot(x=x[:, 0], y=x[:, 1], hue=y)
```

```python
## split the dataset into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
penalty = ['11','12','elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg','lbfgs','liblinear','sag','saga']
class_weight= [{0:w,1:y} for w in [1,10,50,100] for y in [1,10,50,100]]
```

```python
params = [
    {
        "penalty": ["l2"],
        "C": [100, 10, 1.0, 0.1, 0.01],
        "solver": ["newton-cg", "lbfgs", "liblinear", "sag", "saga"],
        "class_weight": class_weight
    },
    {
        "penalty": ["l1"],
        "C": [100, 10, 1.0, 0.1, 0.01],
        "solver": ["liblinear", "saga"],
        "class_weight": class_weight
    },

    {
        "penalty": ["elasticnet"],
        "C": [100, 10, 1.0, 0.1, 0.01],
        "solver": ["saga"],
        "l1_ratio": [0.2, 0.5, 0.8],
        "class_weight": class_weight
    }
]
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid = GridSearchCV(
    estimator=model,
    param_grid=params,
    scoring="accuracy",
    cv=cv,
    n_jobs=-1
)

grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
score = accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

# Logistic Regression With ROC curve and ROC AUC score

```python
#roc curve and auc
from sklearn.datasets import make_classification
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
# generate 2 class datasets
x,y = make_classification(n_samples=1000,n_classes=2,random_state = 1)
```

```python
# split the dataset into train and test
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
# generate a no skill prediction(majority class)
dummy_model_prob = [0 for _ in range(len(y_test))]
dummy_model_prob
```

```python
# basic logistic model
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model_prob = model.predict_proba(x_test)
```

```python
 model_prob = model_prob[:,1]
```

```python
# calculating the score
dummy_model_auc = roc_auc_score(y_test,dummy_model_prob)
model_auc = roc_auc_score(y_test,model_prob)
print(dummy_model_auc)
print(model_auc)
```

```python
# calcilate ROC curves
dummy_fpr,dummy_tpr,_ = roc_curve(y_test,dummy_model_prob)
model_fpr,model_tpr,thresholds = roc_curve(y_test,model_prob)
```

```python
## plotting the roc curve for the model
plt.plot(dummy_fpr,dummy_tpr,linestyle = '--',label = 'Dummy Model')
plt.plot(model_fpr,model_tpr,marker = '.',label = 'Logistic')
## axis labels
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
# show the legend
plt.legend()
# show the plot
plt.show()
```

```python
plt.figure(figsize=(20,50))
plt.plot(dummy_fpr, dummy_tpr, linestyle='--', label='Dummy Model')
plt.plot(model_fpr, model_tpr, linestyle='-', marker='.', label='Logistic')

ax = plt.gca()
for xyz in zip(model_fpr, model_tpr, thresholds):
    ax.annotate('%s' % np.round(xyz[2], 2), xy=(xyz[0], xyz[1]))

plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.legend()
plt.show()
```
